### Connecting to MySQL DataBase

```python
import pymysql
import mysql.connector
from mysql.connector import Error
```

```python
def create_server_connection(host_name, user_name, user_password):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection
pw = "gst4KHC3"
db = "pollution-db" 

connection = create_server_connection("localhost", "root", pw)
```

```python
def create_database(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        print("Database created successfully")
    except Error as err:
        print(f"Error: '{err}'")

create_database_query = "CREATE DATABASE `pollution-db`"
create_database(connection, create_database_query)
```

```python
def create_db_connection(host_name, user_name, user_password, db_name):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password,
            database=db_name
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection
```

```python
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query successful")
    except Error as err:
        print(f"Error: '{err}'")
```

```python
create_station_table = """
CREATE TABLE station (
  `SiteID` INT NOT NULL,
  `Location` VARCHAR(45) NOT NULL UNIQUE,
  `geo_point_2d`VARCHAR(45) NOT NULL,
  PRIMARY KEY (`SiteID`));
 """
connection = create_db_connection("localhost", "root", pw, db) # Connect to the Database
execute_query(connection, create_station_table) # Execute our defined query
```

```python
create_reading_table = """
CREATE TABLE readings (
  `readingID` INT NOT NULL AUTO_INCREMENT,
  `Date Time` DATETIME,
  `NOx` FLOAT,
  `NO2` FLOAT,
  `NO` FLOAT,
  `PM10` FLOAT,
  `PM2.5` FLOAT,
  `VPM2.5` FLOAT,
  `NVPM2.5` FLOAT,
  `CO` FLOAT,
  `O3` FLOAT,
  `SO2` FLOAT,
  `Temperature` FLOAT,
  `RH` FLOAT,
  `Air Pressure` FLOAT,
  `Current` VARCHAR(45),
  `Instrument Type` VARCHAR(45),
  `DateStart` DATETIME,
  `DateEnd` DATETIME,
  `SiteID` INT NOT NULL,
  PRIMARY KEY (`readingID`),
  CONSTRAINT `SiteID`
    FOREIGN KEY (`SiteID`) REFERENCES station (`SiteID`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION);
 """
connection = create_db_connection("localhost", "root", pw, db) # Connect to the Database
execute_query(connection, create_reading_table) # Execute our defined query
```
